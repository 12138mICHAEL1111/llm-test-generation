```python
import matplotlib.pyplot as plt
import numpy as np

# categories = ["GTP4oMini","Claude3Haiku","Gemini1.5Pro","Gemini1.5Flash","GPT3.5Turbo"]
# values = [69.3,68.5,54.9,50.0,41.1]
# categories = ["Claude3Haiku","GTP4oMini","Gemini1.5Pro","GPT3.5Turbo","Gemini1.5Flash"]
# values = [73.7,69.0,56.7,51.3,48.3]

categories = ["Claude3Haiku", "GTP4oMini", "Gemini1.5Pro", "GPT3.5Turbo", "Gemini1.5Flash"]
values1 = [68.5, 69.3, 54.9, 41.1, 50.0]  # First set of values
values2 = [73.7, 69.0, 46.7, 51.3, 48.3]  

categories.reverse()
values1.reverse()
values2.reverse()

bar_width = 0.3
y_positions = np.arange(len(categories))

# Plot bars for both sets of values
plt.barh(y_positions - bar_width/2, values1, height=bar_width, label='1.0')
plt.barh(y_positions + bar_width/2, values2, height=bar_width, label='0.2')

for index, (value1, value2) in enumerate(zip(values1, values2)):
    plt.text(value1, index - bar_width/2, f' {value1}%', va='center', fontsize=8)
    plt.text(value2, index + bar_width/2, f' {value2}%', va='center', fontsize=8)

# Y-ticks and X-ticks formatting
plt.yticks(y_positions, categories, fontsize=10)
plt.xticks(ticks=range(0, 101, 20), labels=[f'{i}%' for i in range(0, 101, 20)])

plt.xlim(0, 85)
plt.savefig("chart/compare_error_2.pdf", format='pdf', bbox_inches='tight')
```
