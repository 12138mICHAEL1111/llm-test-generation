```python
import matplotlib.pyplot as plt
# 数据点
x = [1, 2, 3]
# CR1 = [73.8, 79.2, 81.6]
# CR02 = [75.5,86.2,83.5]

# CPR1=[55.7,55.7,63.7]
# CPR02=[55.0,63.3,64.7]

CR1 = [67.0, 90.6, 90.7]
CR02 = [68.9,86.2,83.5]

CPR1=[55.7,55.7,63.7]
CPR02=[54.9,63.3,64.7]
# 创建图形
plt.figure(figsize=(3.5,3))

plt.plot(x, CR1, label='CP1.0', marker='o',color="black",markersize=3)
plt.plot(x, CR02, linestyle='--', label='CP0.2',marker='o',color="black",markersize=3)

plt.plot(x, CPR1, label='CPR1.0', marker='o',color="red",markersize=3)
plt.plot(x, CPR02, linestyle='--', label='CPR0.2',marker='o',color="red",markersize=3)

plt.xticks(x)


# 设置图表标题和坐标轴标签
plt.title('')
plt.xlabel('Level')
plt.ylabel('Percentage')
plt.tick_params(axis='both', labelsize='small') 

# 设置纵轴的范围
plt.ylim(30, 100)

# 添加图例
plt.legend(fontsize="small")

plt.savefig("chart/gemini/cp&cpr.pdf", format='pdf', bbox_inches='tight')

```

```python
import matplotlib.pyplot as plt
import numpy as np

# 示例数据
categories = ['C.P', 'C.P.R', 'Cov', 'Mu']
GPT = [81.3, 60.8, 38.1, 30.5]
Gemini = [83.9,63.3,47.5,37.8]

# 每个柱的位置
x = np.arange(len(categories))  # 标签位置
width = 0.2  # 柱的宽度

# 创建图形
plt.figure(figsize=(4, 3))

# 绘制柱状图
plt.bar(x - width/2, GPT, width, label='ChatGPT3.5', color='blue')
plt.bar(x + width/2, Gemini, width, label='Gemini1.5PRO', color='green')

# 添加标题和轴标签
plt.title('Comparison of ChatGPT and Gemini')
plt.xlabel('Metrics')
plt.ylabel('Pecentage')

# 添加类别名称
plt.xticks(x, categories)

# 添加图例
plt.legend(fontsize="small")

# 显示图形
plt.savefig("chart/compare.pdf", format='pdf', bbox_inches='tight')
```
