```python
import matplotlib.pyplot as plt
import numpy as np

# 示例数据
categories = ['C.P', 'C.P.R', 'Cov', 'Mu']
GPT = [81.3, 60.8, 37.2, 29.2]
Gemini = [83.9,63.3,47.3,37.7]

# 每个柱的位置
x = np.arange(len(categories))  # 标签位置
width = 0.2  # 柱的宽度

# 创建图形
plt.figure(figsize=(4, 3))

# 绘制柱状图
plt.bar(x - width/2, GPT, width, label='GPT3.5Turbo', color='blue')
plt.bar(x + width/2, Gemini, width, label='Gemini1.5Pro', color='green')

# 添加标题和轴标签
plt.title('LLMs Comparison of Test Generation')
plt.xlabel('Metrics')
plt.ylabel('Score')

# 添加类别名称
plt.xticks(x, categories)

# 添加图例
plt.legend(fontsize="small")

# 显示图形
plt.savefig("chart/compare_test.pdf", format='pdf', bbox_inches='tight')
```
